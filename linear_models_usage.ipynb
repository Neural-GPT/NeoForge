## **Linear Regression**

```python
from linear_models import Linear_Regression
import pandas as pd
```

```python
x = pd.DataFrame({"F1":[1,2,3,4,1], "F2":[3,4,3,1,5], "F3": [0,1,5,3,9], "F4": [6,0,1,2,5]})
y = pd.Series([28,31,42,38,50]) # Y = 5F1 + 3F2 + 2F3 + F4 + 7
x
```

```python
model = Linear_Regression(l2 = 0, alpha = 0.01, iter = 200)
model.fit(x, y, freeze = 0.1, random_state = 2,  verbose = 10, history_save_interval = 10, cosine_annealing = True, normalise = True)
```

```python
model.predict(x)
```

```python
model.plot_cost()
```

```python
model.coefficient_path()
```

```python
model.coefficients_
```

```python
model.intercept_
```

```python
model.feature_importance()
```

```python
model.lr_plot()
```

```python
model.summary()
```
